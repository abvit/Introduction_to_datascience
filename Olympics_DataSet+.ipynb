```python
#Loading the olympics dataset which was derived from the Wikipedia data on All Time Olympic Games Medals
import pandas as pd
df = pd.read_csv('olympics.csv',index_col = 0,skiprows = 1)
df.head()
```

```python
#To drop the row Totals
df= df.drop('Totals')
```

```python
#To split the country name and initials
# '\' indicates regular expressions are used for matching 
name_ids = df.index.str.split('\s\(')
```

```python
df.index = name_ids.str[0]
```

```python
name_ids.str[1].str[0:3]
```

```python
#Creating a new column to store ids
df['ID'] = name_ids.str[1].str[0:3]
```

```python
df.head()
```

```python
#To convert 01!,02!,03! to Gold,Silver,Bronze
#When inplace=True is passed then the data is renamed in the place and returns nothing:
#df.an_operation(inplace= True)
#When inplace = False is passed(ie it is default), the operation is performed and the copy of the object is passed:
#df = df.anoperation(inplace=false)

for col in df.columns:
    if col[0:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]} , inplace = True)
    if col[0:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace = True)
    if col[0:2] == '03':
        df.rename(columns={col:'Bronze'+col[4:]},inplace = True)
    if col[0:1]=='№':
        df.rename(columns={col:'#'+ col[1:]},inplace = True)
df.head()
        
```
